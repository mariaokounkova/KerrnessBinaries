```python
def InitialConstraints():
    
    plt.figure(figsize=(8,5))
    ham, mom = np.loadtxt("Data/MultipolarID/AdmEqns.dat", comments="#",usecols=([1,2]),unpack=True)
    res = [(9+4*k)*(5+2*k) for k in range(len(ham))]
    plt.plot(res, ham, '-o', label='Hamiltonian', color='#56d5e3')
    plt.plot(res, mom, '--o', label='Momentum', color='#f08bb0')
    
    plt.yscale('log')
    plt.legend(frameon=False)
    plt.ylabel('$\|$ Constraint Norm $\|_\mathrm{L2}$')
    plt.xlabel(r'$N_R \times N_L$')
    plt.show()
    

InitialConstraints()
```
